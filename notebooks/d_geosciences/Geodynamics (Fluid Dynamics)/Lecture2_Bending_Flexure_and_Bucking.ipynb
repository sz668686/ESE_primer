(Lecture 2)

## Application to Lithospheric Plates

Consider the oceanic crust with thickness $h$ and density $\rho_w$, lying over the oceanic lithosphere with density $\rho_m$, which in turn lies over the fluid mantle with density $\rho_m$. We define the deformation $w(x)$ after deformation, and the net upwards buoyant force $g(x)$ is given by:

$$
g(x) = -(\rho_m - \rho_w)w(x)
$$

If $q(x)$ is the additional applied load, then:

### For Oceanic Crust:

$$
D w''''(x) = g(x) - (\rho_m - \rho_w) g w(x)
$$

### For Continental Crust:

$$
D w''''(x) = g(x) - (\rho_a - \rho_c) g w(x)
$$

### Example: Deformation Due to the Weight of a Mountain Chain

For a periodic mountain chain with amplitude $h_0$ and wavelength $\lambda$, $g(x)$ can be modeled as:

$$
g(x) = \rho_c g h_0 \sin \left(\frac{2\pi x}{\lambda}\right)
$$

### Solving the Differential Equation

Given the ordinary differential equation (ODE):

$$
D w''''(x) + (\rho_m - \rho_c)g w(x) = \rho_c g h_0 \sin \left(\frac{2\pi x}{\lambda}\right)
$$

Assume a solution having the same wavelength $\lambda$, but an as-yet unknown amplitude, $w_0$:

$$
w(x) = w_0 \sin \left(\frac{2\pi x}{\lambda}\right)
$$

Then, the fourth derivative of $w(x)$ is:

$$
w''''(x) = \left(\frac{2\pi}{\lambda}\right)^4 w_0 \sin \left(\frac{2\pi x}{\lambda}\right)
$$

### Inserting into the ODE:

$$
D \left(\frac{2\pi}{\lambda}\right)^4 w_0 \sin \left(\frac{2\pi x}{\lambda}\right) + (\rho_m - \rho_c)g w_0 \sin \left(\frac{2\pi x}{\lambda}\right) = \rho_c g h_0 \sin \left(\frac{2\pi x}{\lambda}\right)
$$

Cancel out the common term $\sin \left(\frac{2\pi x}{\lambda}\right)$:

$$
D \left(\frac{2\pi}{\lambda}\right)^4 w_0 + (\rho_m - \rho_c)g w_0 = \rho_c g h_0
$$

Solve for $w_0$:

$$
w_0 = \frac{\rho_c g h_0}{D \left(\frac{2\pi}{\lambda}\right)^4 + (\rho_m - \rho_c)g}
$$

This gives the amplitude of the deformation.

### Normalized Deflection

Define the non-dimensional parameter $\chi^*$ as:

$$
\chi^* = \left(\frac{D}{\rho g h}\right)^{1/4}
$$

Then:

$$
\delta^* = \frac{w_0}{h_0} = \frac{\rho_c}{(\rho_m - \rho_c)} \frac{1}{1 + \left(\frac{\chi^*}{\lambda}\right)^4}
$$

For small $\lambda$ (i.e., $\lambda < 0.5\lambda^*$):

$$
\frac{w_0 (\rho_m - \rho_c)}{h_0 \rho_c} \approx \left(\frac{\lambda}{\chi^*}\right)^4 \approx \lambda^4 < 1
$$

For large $\lambda$ (i.e., $\lambda > 2\lambda^*$):

$$
\frac{w_0 (\rho_m - \rho_c)}{h_0 \rho_c} \approx 1
$$

Typical values of $\lambda^*$ are around 400 km.

## Bending of Lithosphere Under an Island Chain

The crust and mantle deformation under an island chain load can be described as follows:

**Crust** $ \longrightarrow w(x) \longleftarrow $ **Mantle**

*Note: Islands form a chain into a facet; load is assumed to be localized in a narrow region centered around $x = 0$.*

The ordinary differential equation (ODE) is given by:

$$
D w''''(x) + E w''(x) + \Delta \rho g w(x) = 0
$$

Setting the axial load $P = 0$ (we will consider the effect of $P \neq 0$ later) and due to the symmetry of the problem, we can solve in the region $0 < x < \infty$.

Thus, we have:

$$
w''''(x) + \left(\frac{\Delta \rho g}{D}\right) w(x) = 0
$$

We need 4 boundary conditions; 2 at $x = 0$, 2 at $x = \infty$:

1. Assuming symmetry and continuity at $x = 0$ (i.e., the crust does not crack), then $w'(x) = 0$ must be zero.
2. If the total weight of islands is $W_0$, then each half of the plate must hold up half the load. Therefore, $V(x = 0) = \frac{W_0}{2}$.
3. At $x = \infty$, we must have finite values of $w(x)$ and $w'(x)$.

### Solving the Linear ODE with Constant Coefficients

To solve the linear ordinary differential equation (ODE) with constant coefficients, we assume a solution of the form $w(x) = C e^{mx}$. The fourth derivative $w''''(x)$ is then:

$$
w''''(x) = C m^4 e^{mx}
$$

Inserting this into the ODE gives:

$$
C m^4 e^{mx} + \left(\frac{\Delta \rho g}{D}\right) C e^{mx} = 0
$$

Factoring out $C e^{mx}$:

$$
C e^{mx} \left[m^4 + \frac{\Delta \rho g}{D}\right] = 0 \Rightarrow m = \left(\frac{-\Delta \rho g}{D}\right)^{1/4}
$$

### Note: The fourth-order ODE implies there may be 4 roots for $m$:

$$
m = \left(\frac{-\Delta \rho g}{D}\right)^{1/4} = (-1)^{1/4} \left(\frac{\Delta \rho g}{D}\right)^{1/4}
$$

Recalling that in the complex plane, $-1$ can be written as $e^{i\pi}$ (i.e., $z = re^{i\theta}$ where $r = 1$, $\theta = \pi$), we have:

$$
-1 = e^{i\pi} = e^{i(\pi + 2n\pi)} \text{ for all } n
$$

So:

$$
(-1)^{1/4} = \left(e^{i(\pi + 2n\pi)}\right)^{1/4}
$$

### Complex Roots and General Solution

We can then illustrate the four roots of $-1$ in the complex plane. So, $m$ is given by:

$$
m = \left(\frac{\Delta \rho g}{D}\right)^{1/4} \left(i^{1/4}, -i^{1/4}, i^{-1/4}, -i^{-1/4}\right) = \left(\frac{\Delta \rho g}{D}\right)^{1/4} \frac{1}{\sqrt{2}} \left(1 + i, -1 + i, -1 - i, 1 - i\right)
$$

Call this $\frac{1}{\sqrt{2}} \left(1 + i\right)k$ where $k$ is the flexural parameter. Then the general solution for $w(x)$ is:

$$
w(x) = C_1 e^{(1+i)kx} + C_2 e^{-(1+i)kx} + C_3 e^{(1-i)kx} + C_4 e^{-(1-i)kx}
$$

Expanding and grouping real and imaginary parts, we get:

$$
w(x) = e^{kx/\sqrt{2}} \left[ C_1 e^{ikx/\sqrt{2}} + C_2 e^{-ikx/\sqrt{2}} + C_3 e^{ikx/\sqrt{2}} + C_4 e^{-ikx/\sqrt{2}} \right]
$$

$$
= e^{kx/\sqrt{2}} \left[ (C_1 + C_3) \cos \left(\frac{kx}{\sqrt{2}}\right) + i(C_1 - C_3) \sin \left(\frac{kx}{\sqrt{2}}\right) \right] + e^{-kx/\sqrt{2}} \left[ (C_2 + C_4) \cos \left(\frac{kx}{\sqrt{2}}\right) + i(C_2 - C_4) \sin \left(\frac{kx}{\sqrt{2}}\right) \right]
$$

Redefine constants as $A_1$, $A_2$, $A_3$, $A_4$ for simplicity, resulting in:

$$
w(x) = e^{kx/\sqrt{2}} \left[A_1 \cos \left(\frac{kx}{\sqrt{2}}\right) + A_2 \sin \left(\frac{kx}{\sqrt{2}}\right)\right] + e^{-kx/\sqrt{2}} \left[A_3 \cos \left(\frac{kx}{\sqrt{2}}\right) + A_4 \sin \left(\frac{kx}{\sqrt{2}}\right)\right]
$$

This is the general solution. Now we proceed to look at boundary conditions.


## Boundary Conditions and Solution Continuity

For $w(x)$ to be bounded (i.e., finite as $x \to \infty$), we need $A_1 = A_2 = 0$. Thus far we know that:

$$
w(x) = e^{-kx}\left(A_3 \cos kx + A_4 \sin kx\right)
$$

Assuming symmetry and continuity at $x = 0$, we must have $w'(0) = 0$. Thus,

$$
w'(x) = -\frac{k}{2} e^{-kx/2} \left( 2A_3 \cos \frac{kx}{2} + A_4 \sin \frac{kx}{2} \right) + \frac{k}{2} e^{-kx/2} \left(-A_3 \sin \frac{kx}{2} + A_4 \cos \frac{kx}{2}\right)
$$

So, at $x = 0$, $w'(0)$ must be zero, which implies $A_4 = A_3$. Hence,

$$
w(x) = A_3 e^{-kx} \left(\cos kx + \sin kx\right)
$$

Lastly, we must use the fact that at $x = 0$, $V(x = 0) = -\frac{W_0}{2}$. Recalling our list of five governing equations, particularly:

- **Equation 3:** $M'(x) = V(x) + p w'(x)$, but $p = 0$ here, so $M'(x) = V(x)$
- **Equation 5:** $M(x) = -D w''(x)$, so $M'(x) = -D w'''(x)$

Therefore, at $x = 0$, $w'''(0) = \frac{V_0}{2D}$, and we can compute $w'''(x)$ from $w(x)$ and plug in $x = 0$ to find $A_3$.

## Deflection Analysis

The deflection is given by:

$$
w(x) = \frac{V_0 x^3}{8D} e^{-x/\alpha} \left(\cos \frac{x}{\alpha} + \sin \frac{x}{\alpha}\right), \quad \text{for } x \geq 0
$$

At $x = 0$, the deflection $w$ is maximum, denoted as $w_0$:

$$
w_0 = \frac{V_0 \alpha^3}{8D}
$$

So, the normalized deflection is:

$$
\frac{w(x)}{w_0} = e^{-x/\alpha} \left(\cos \frac{x}{\alpha} + \sin \frac{x}{\alpha}\right), \quad \text{for } x \geq 0
$$

The half-width of the depression, labeled as $X_b$, can be found by setting $w(X_b) = 0$:

$$
\tan \frac{X_b}{\alpha} = -1
$$

$$
\frac{X_b}{\alpha} = \arctan(-1) = 135^\circ = \frac{3\pi}{4}
$$

Using the typical values:

$$
E = 100 \, \text{GPa}, \, h = 50 \, \text{km}, \, \nu = 0.25, \, \rho_m = 3300 \, \text{kg/m}^3, \, \rho_w = 1000 \, \text{kg/m}^3
$$

$$
\alpha \approx 100 \, \text{km}, \, X_0 \approx 250 \, \text{km}
$$

## Analysis of Deflection

Note that for $x > x_0$, the plate actually bulges upwards. To find the location of this "bulge," we look for the first value of $x > 0$ for which $w'(x) = 0$ (i.e., local maxima).

Given $w(x)$ as:

$$
w(x) = w_0 e^{-x/\alpha} \left(\cos \frac{x}{\alpha} + \sin \frac{x}{\alpha}\right), \quad x \geq 0
$$

The first derivative $w'(x)$ is:

$$
w'(x) = w_0 \frac{d}{dx} \left[e^{-x/\alpha} \left(\cos \frac{x}{\alpha} + \sin \frac{x}{\alpha}\right)\right]
$$

$$
= w_0 e^{-x/\alpha} \left(-\frac{1}{\alpha} \cos \frac{x}{\alpha} - \frac{1}{\alpha} \sin \frac{x}{\alpha} + \frac{1}{\alpha} \sin \frac{x}{\alpha} - \frac{1}{\alpha} \cos \frac{x}{\alpha}\right)
$$

$$
= -\frac{2w_0}{\alpha} e^{-x/\alpha} \sin \frac{x}{\alpha}
$$

Setting $w'(x) = 0$ gives the condition for the extrema:

$$
\sin \frac{x}{\alpha} = 0 \Rightarrow \frac{x}{\alpha} = n\pi, \, n = 0, 1, 2, \dots
$$

The first local maximum occurs at $x_1 = \pi \alpha$.

The height of this first bulge can be found by evaluating $w(x)$ at $x_1$:

$$
w(x_1) = w(\pi \alpha) = w_0 e^{-\pi} (\cos \pi + \sin \pi) = -w_0 e^{-\pi}
$$

Using the typical values given:

$$
E = 100 \, \text{GPa}, \, \nu = 0.25, \, \rho_m = 3300 \, \text{kg/m}^3, \, \rho_w = 1000 \, \text{kg/m}^3
$$

Thus, $\alpha \approx 100 \, \text{km}$, and $x_0 \approx 250 \, \text{km}$.
