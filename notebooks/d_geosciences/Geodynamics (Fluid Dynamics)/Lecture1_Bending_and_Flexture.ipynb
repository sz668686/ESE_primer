(Lecture 1)

# Learning Outcomes

1. Understand the equation governing the bending of a beam and solve it.
2. Understand the equations used as models for geological structures.

## Beam and Plate Theory

- A **beam** is long in one direction and thin in another.
- A **plate** is a beam extended into the third dimension.
- **Mechanical reasons**: Plates that are very thin, such as lithospheric plates, satisfy plate theory.

## Derivation of the Equation

Our goal is to find the deformation of a plate, denoted $w$, and derive an ordinary differential equation (ODE) for $w$.

- Consider a **cartoon of a plate**: derive from beam to 3D.
- Use $x$ and $y$ coordinates, with $y$ pointing downwards.
- $w(x)$ represents the **vertical displacement downward**.
- $q(x)$ is the **load exerted on the beam** in the vertical direction.
- We assume a unit depth into the $z$ direction.

## Free Body Diagram

1. Set up the **coordinate system**.
2. Identify all **forces** acting on the material.
3. **Lateral force** $p$ acts in the horizontal $x$ direction.
4. **Shear force** $V$ acts in the vertical $y$ direction.
5. The force and **moments** $M(x)$ and $M(x + \Delta x)$ are considered.
6. Apply **Newton’s Laws** to derive the necessary equations.

## Conclusion

- **Newton’s law** in the $y$ direction gives us the differential equation where the rate of shear stress variation with $x$ is equal to $-q(x)$:

$$
\frac{dV}{dx} = -q(x)
$$

- To obtain the ODE of $w$, we replace $\frac{dV}{dx} = -q(x)$ into the equation and relate the moment $M(x)$ to the displacement $w(x)$.


## Apply Newton’s Law on the Beam

The equilibrium conditions for the forces in the $x$ and $y$ directions are given by:

$$
\sum F_x = 0 \Rightarrow P(x) - [P(x + \Delta x)] = 0 \Rightarrow P(x + \Delta x) = P(x)
$$

$$
\Rightarrow P(x) \text{ is constant and denoted as } P.
$$

$$
\sum F_y = 0 \Rightarrow q(x) \Delta x + V(x + \Delta x) - V(x) = 0
$$

$$
\Rightarrow V(x + \Delta x) - V(x) = -q(x) \Delta x.
$$

Taking the limit as $\Delta x \to 0$, we obtain the differential relationship:

$$
\frac{dV}{dx} = -q(x).
$$

### Sum of the Moments Equals Zero

Considering the moments around the left edge of a differential beam element, we have the sum of the moments equals zero:

$$
\sum M \, \text{(ccw about left edge)} = 0
$$

This leads to the following equation:

$$
M(x + \Delta x) - M(x) - V(x + \Delta x)\Delta x - P(x + \Delta x)\Delta w = 0
$$

Dividing through by $\Delta x$:

$$
\frac{M(x + \Delta x) - M(x)}{\Delta x} - P \frac{\Delta w}{\Delta x} = V(x + \Delta x)
$$

Taking the limit as $\Delta x \to 0$:

$$
\frac{dM}{dx} - P \frac{dw}{dx} = V(x)
$$

To eliminate $V(x)$, we differentiate and recall that $\frac{dV}{dx} = -q(x)$:

$$
\frac{d^2M}{dx^2} = P \frac{d^2w}{dx^2} + \frac{dV}{dx} = P \frac{d^2w}{dx^2} - q(x)
$$

Now, we need to relate $M$ to $w$.

### Consider Stress and Strain in a Deformed Beam

Axial strain $\epsilon_{xx}$ is given by:

$$
\epsilon_{xx} = \frac{\text{Old length} - \text{New length}}{\text{Old length}}
$$

Consider a certain value of $y$:

- Old length: $R \Delta \phi$ (i.e., length of neutral axis)
- New length: $(R - y) \Delta \phi$

So,

$$
\epsilon_{xx} = \frac{R \Delta \phi - (R - y) \Delta \phi}{R \Delta \phi} = \frac{y}{R}
$$

Then, by Hooke’s law:

$$
\sigma_{xx} = E \epsilon_{xx} = \frac{E y}{R}
$$

### Moment Calculation for a Deformed Beam

For regions above and below the neutral axis $y = 0$, the stress $\sigma_{xx}$ due to bending is given by:

$$
\sigma_{xx} = \frac{E y}{R}
$$

where $E$ is Young’s modulus, $y$ is the vertical position from the neutral axis, and $R$ is the radius of curvature of the beam.

- For $y > 0$ (compression): $\sigma_{xx} > 0$
- For $y < 0$ (tension): $\sigma_{xx} < 0$

Assuming a unit thickness of the beam, the differential force $dF$ is:

$$
dF = \sigma(y) dy
$$

The differential moment $dM$ at a distance $y$ from the neutral axis is:

$$
dM = y dF = y \sigma(y) dy
$$

Integrating over the height of the beam to find the total bending moment $M$:

$$
M = \frac{E}{R} \int_{-\frac{h}{2}}^{\frac{h}{2}} y^2 dy = \frac{E}{R} \left[ \frac{y^3}{3} \right]_{-\frac{h}{2}}^{\frac{h}{2}}
$$

$$
M = \frac{E}{R} \left( \frac{h^3}{8} - \left(-\frac{h^3}{8}\right) \right) = \frac{E h^3}{12R}
$$

Therefore, the bending moment $M(x)$ is:

$$
M(x) = \frac{E h^3}{12R}
$$

Insert $M(x) = \frac{E h^3}{12R(x)}$ into $\frac{d^2M}{dx^2} = p\frac{d^2w}{dx^2} - q(x)$:

$$
\frac{d^2}{dx^2} \left[ \frac{E h^3}{12R(x)} \right] = p \frac{d^2w}{dx^2} - q(x)
$$

Recall from calculus:

$$
\frac{1}{R} = -\frac{d^2w}{dx^2}
$$

So:

$$
\frac{d^2}{dx^2} \left[ -\frac{E h^3}{12} \frac{d^2w}{dx^2} \right] = p \frac{d^2w}{dx^2} - q(x)
$$

$$
\frac{E h^3}{12} \frac{d^4w}{dx^4} + p \frac{d^2w}{dx^2} = q(x)
$$

For a plate that extends in the 3-D direction:

$$
\frac{E h^3}{12(1 - \nu^2)} \frac{d^4w}{dx^4} + p \frac{d^2w}{dx^2} = q(x) \quad \text{(Sophie Germain, 1801)}
$$

where $12(1 - \nu^2)$ is the **Flexure Rigidity** $D$, and $\nu$ is **Poisson's Ratio**.


## Stress-Strain Relationship in Bending

When a beam made of an elastic material, such as steel or rock, is bent, it experiences both compression and tension:

- The top surface ($y = \frac{h}{2}$) undergoes compression, which shortens its length.
- The bottom surface ($y = -\frac{h}{2}$) is under tension, which lengthens it.
- Convert stress to force: $\text{Force} = \text{Stress} \times \text{Area}$

The strain in the beam, defined as the fractional change in length, is given by:

$$
\text{Strain} = \frac{\text{Old length} - \text{New length}}{\text{Old length}} = \frac{\text{Arc length at } y - (\text{Radius} - y) \times \text{angle}}{\text{Arc length at } y}
$$

Arc length at $y$ is given by:

$$
\text{Arc length at } y = \frac{y}{R}
$$

where $R$ is the radius of curvature of the bent beam. The stress at any point in the beam can then be calculated using Hooke’s law:

$$
\text{Stress} = E \times \text{Strain} = E \times \frac{y}{R}
$$

where $E$ is Young’s modulus of the material. The stress distribution is linear across the height of the beam, with compression at the top and tension at the bottom.

## Calculating the Bending Moment

The bending moment $M(x)$ of the beam is calculated by integrating the stress over the cross-sectional area:

$$
M(x) = \int_{-\frac{h}{2}}^{\frac{h}{2}} \text{Stress} \times y \, dA
$$

The differential area $dA$ is simply the width of the beam times the differential height element $(b \, dy)$. By integrating the moment of the stress distribution about the neutral axis, we obtain the bending moment.

## Flexural Rigidity

The beam’s resistance to bending is quantified by the flexural rigidity $D$, defined as:

$$
D = E \times I
$$

where $I$ is the second moment of area (also known as the moment of inertia) of the beam’s cross section. This parameter is central to the plate bending theory as it relates the material properties and geometry to the bending behavior.

## Summary of Useful Equations for Plate Bending

1. **The plate bending equation:**

$$
D \frac{d^4w}{dx^4} + p \frac{d^2w}{dx^2} = q(x)
$$

2. **The shear force derivative:**

$$
\frac{dV}{dx} = -q(x)
$$

3. **The relationship between moment and shear force:**

$$
\frac{dM}{dx} = V + p \frac{dw}{dx}
$$

4. **The moment curvature relationship:**

$$
\frac{d^2M}{dx^2} = -q(x) + p \frac{d^2w}{dx^2}
$$

5. **The moment in terms of the deflection $w$:**

$$
M = -D \frac{d^2w}{dx^2}
$$
